## 4. Make an assumption about a distribution law (a hypothesis H0 the sample came from a normal distribution law)

```python
import numpy as np
from scipy.stats import norm, chi2

# Basic statistics of the data
mean_value = np.mean(data)
standard_deviation = np.std(data)

# Define the intervals
bins = [60, 65, 70, 75, 80, 85, 90, 95, 100]

# Calculate the theoretical probabilities for each interval (using the CDF of the normal distribution)
theoretical_probabilities = norm.cdf(bins[1:], mean_value, standard_deviation) - norm.cdf(bins[:-1], mean_value, standard_deviation)
theoretical_frequencies = theoretical_probabilities * len(data)  # Multiply by the total sample size to get expected frequencies

# Calculate the observed frequencies
observed_frequencies, _ = np.histogram(data, bins=bins)

# Print the calculated results
print("Theoretical Frequencies:", theoretical_frequencies)
print("Observed Frequencies:", observed_frequencies)

# Calculate the chi-squared statistic
chi_squared_stat = np.sum((observed_frequencies - theoretical_frequencies) ** 2 / theoretical_frequencies)
degrees_of_freedom = len(bins) - 1 - 2  # Number of intervals - 1 - number of parameters (mean and standard deviation)

# Print the chi-squared statistic and degrees of freedom
print("Chi-squared Statistic:", chi_squared_stat)
print("Degrees of Freedom:", degrees_of_freedom)

# Calculate the chi-squared critical value (significance level = 0.05)
chi_squared_critical = chi2.ppf(0.95, degrees_of_freedom)

# Test result
if chi_squared_stat < chi_squared_critical:
    print("Accept the null hypothesis: the sample comes from a normal distribution")
else:
    print("Reject the null hypothesis: the sample does not come from a normal distribution")
```

```python
import pandas as pd
import numpy as np

# Reconstruct table data
table_data = {
    "Intervals": [f"{int(bins[i])}-{int(bins[i+1])}" for i in range(len(bins)-1)],
    "Observed Frequencies (Oi)": observed_frequencies,
    "Expected Frequencies (Ei)": theoretical_frequencies,
    "Chi-Square Component": (observed_frequencies - theoretical_frequencies) ** 2 / theoretical_frequencies,
    "Standardized Residual": (observed_frequencies - theoretical_frequencies) / np.sqrt(theoretical_frequencies)
}

# Create DataFrame to display the table
chi_square_table = pd.DataFrame(table_data)
total_chi_square_statistic = chi_square_table["Chi-Square Component"].sum()

# Add total chi-square statistic
chi_square_table.loc["Total"] = ["Total", chi_square_table["Observed Frequencies (Oi)"].sum(), chi_square_table["Expected Frequencies (Ei)"].sum(), total_chi_square_statistic, chi_square_table["Standardized Residual"].sum()]

chi_square_table
```

## 6. Check this assumption using Romanovskiy’s test and test ${X^2}$ for goodness of fit.

```python
# Calculate standardized residuals
chi_square_table["Standardized Residual"] = (chi_square_table["Observed Frequencies (Oi)"] - chi_square_table["Expected Frequencies (Ei)"]) / np.sqrt(chi_square_table["Expected Frequencies (Ei)"])

# Calculate the sum of squared standardized residuals
sum_of_squared_residuals = np.sum(chi_square_table["Standardized Residual"] ** 2)

# Print the result
print("Sum of Squared Standardized Residuals:", sum_of_squared_residuals)

# Romanovskiy test result
if sum_of_squared_residuals < chi_squared_critical:
    print("Accept the null hypothesis: the sample comes from a normal distribution")
else:
    print("Reject the null hypothesis: the sample does not come from a normal distribution")

chi_square_table
```

### Romanovskiy Test Result
- Sum of squares of the standardized residuals: 34.33
- The critical value of $X^2$ : 12.59
Since the sum of squares of the standardized residuals (34.33) is greater than the ${X^2}$critical value (12.59), we reject the null hypothesis that the sample does not come from a normal distribution.
