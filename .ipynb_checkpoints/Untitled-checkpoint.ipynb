## 1. Construct a continuous statistic series

```python
import pandas as pd

# 提供的数据值
data = [
    85, 77, 63, 88, 64, 61, 70, 71, 85, 68,
    81, 70, 86, 72, 90, 93, 79, 66, 84, 64,
    73, 83, 78, 79, 73, 81, 91, 92, 73, 68,
    88, 80, 76, 100, 67, 93, 81, 83, 76, 70,
    100, 90, 83, 75, 79, 66, 91, 95, 78, 93,
    92, 69, 97, 97, 75, 89, 82, 72, 72, 79,
    99, 77, 96, 70, 69, 89, 95, 91, 96, 77,
    75, 81, 75, 63, 93, 90, 97, 68, 75, 97,
    89, 97, 61, 67, 84, 89, 83, 89, 75, 97,
    100, 87, 81, 81, 93, 79, 100, 97, 78, 93
]

# 排序数据
sorted_data = sorted(data)

# 计算频率
from collections import Counter
frequency = Counter(sorted_data)

# 转换为列表以便展示
sorted_data_with_frequency = [(value, frequency[value]) for value in sorted(set(sorted_data))]
# 创建 DataFrame 来更美观地展示数据
data_df = pd.DataFrame(sorted_data_with_frequency, columns=['value', 'frequency'])
data_df
```

## 2. Find the numerical сharacteristics 3. Plot a histogram

```python
# 计算均值、方差、标准差
mean_value = data_df['value'].mean()
variance = data_df['value'].var()
standard_deviation = data_df['value'].std()

# 计算中位数和众数
median_value = data_df['value'].median()
mode_value = data_df['value'].mode()[0]  # 取第一个众数

# 计算数据的最小值和最大值
min_value = data_df['value'].min()
max_value = data_df['value'].max()


# Printing the results
print("Mean:", mean_value)
print("Variance:", variance)
print("Standard Deviation:", standard_deviation)
print("Median:", median_value)
print("Mode:", mode_value)
print("Minimum Value:", min_value)
print("Maximum Value:", max_value)
```

## Plot a histogram

```python
import matplotlib.pyplot as plt

# 从原始数据创建直方图
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## 4. Make an assumption about a distribution law (a hypothesis H0 the sample came from a normal distribution law)

```python
import numpy as np
from scipy.stats import norm

# 确定区间的边界（与直方图相同）
bins = np.linspace(min_value, max_value, 21)  # 使用20个区间
bin_centers = 0.5 * (bins[:-1] + bins[1:])  # 计算区间中点

# 计算每个区间的理论概率（使用正态分布的累积分布函数）
theoretical_probabilities = norm.cdf(bins[1:], mean_value, standard_deviation) - norm.cdf(bins[:-1], mean_value, standard_deviation)
theoretical_frequencies = theoretical_probabilities * len(data)  # 乘以样本总数得到期望频率

# 将结果与实际观测值进行比较（绘制直方图和理论曲线）
observed_frequencies, _ = np.histogram(data, bins=bins)

# 打印计算结果
theoretical_frequencies, observed_frequencies
```

```python
# 计算卡方统计量
chi_squared_stat = ((observed_frequencies - theoretical_frequencies) ** 2 / theoretical_frequencies).sum()

# 打印计算的卡方统计量
chi_squared_stat
```

## Fulfill the table from X^2

```python
# 计算罗马诺夫斯基值
romanovsky_values = (observed_frequencies - theoretical_frequencies) / np.sqrt(theoretical_frequencies)

# 打印罗马诺夫斯基值
romanovsky_values
```

```python
# 我们需要调整区间的数量以匹配我们的数据点数
# 取得正确的数据点数目
num_intervals = len(observed_frequencies)

# 匹配区间和数据点
intervals = [f"{int(bins[i])}-{int(bins[i+1])}" for i in range(num_intervals)]

# 重新构建表格数据
table_data = {
    "Interval": intervals,
    "Observed Frequencies (Oi)": observed_frequencies,
    "Expected Frequencies (Ei)": theoretical_frequencies,
    "Chi-Square Component": (observed_frequencies - theoretical_frequencies) ** 2 / theoretical_frequencies
}

# 创建 DataFrame 来展示这个表格
chi_square_table = pd.DataFrame(table_data)
total_chi_square_statistic = chi_square_table["Chi-Square Component"].sum()

# 添加总卡方统计量
chi_square_table.loc["Total"] = ["Total", chi_square_table["Observed Frequencies (Oi)"].sum(), chi_square_table["Expected Frequencies (Ei)"].sum(), total_chi_square_statistic]

chi_square_table
```
